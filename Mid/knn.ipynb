```python
import os
import random
import cv2
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_DIR = 'C:/Users/muvin/OneDrive/Desktop/CIFAR-10-images-master/train'
CATEGORIES = []
for c in os.listdir(TRAIN_DIR):
    CATEGORIES.append(c)
print(CATEGORIES)
```

```python
TRAIN_DATA = []
for c in CATEGORIES:
    path = os.path.join(TRAIN_DIR,c)
    class_num = CATEGORIES.index(c)
    for img in tqdm(os.listdir(path)):
        img_arr = cv2.imread(os.path.join(path,img))
        TRAIN_DATA.append([img_arr, class_num])
print(len(TRAIN_DATA))
```

```python
random.shuffle(TRAIN_DATA)
plt.figure(figsize=(20,10))

for i in range(50):
    plt.subplot(5,10,i+1)
    plt.imshow(TRAIN_DATA[i][0])
    plt.xlabel(CATEGORIES[TRAIN_DATA[i][1]])
    plt.xticks([])
    plt.yticks([])
    if i==50:
        break
plt.show()
```

```python
f0 = TRAIN_DATA[0:1000]
f1 = TRAIN_DATA[1000:2000]
f2 = TRAIN_DATA[2000:3000]
f3 = TRAIN_DATA[3000:4000]
f4 = TRAIN_DATA[4000:5000]
```

```python
import math 

center1=[]
center2=[]
K=[2,3,4,6,9]

for i in range(5):
    sum_acc1 = []
    sum_acc2 = []
    
    if i==0:
        train = f1+f2+f3+f4
        validation = f0 
    elif i==1:
        train = f0+f2+f3+f4
        validation = f1
    elif i==2:
        train = f1+f0+f3+f4
        validation = f2
    elif i==3:
        train = f1+f2+f0+f4
        validation = f3
    elif i==4:
        train = f1+f2+f3+f0
        validation = f4
        

    for k in K:
        acc1=0
        acc2=0
        for v in tqdm(validation):
            number1 = []
            number2 = []
            for t in train:
                distance_2 = np.sum(np.sqrt(t[0]**2-v[0]**2))
                number1.append([distance_2,t[1]])

                distance_1 = np.sum(t[0]-v[0])
                number2.append([distance_1,t[1]])
            sorted_list1 = sorted(number1, key=lambda a: a[0])
            sorted_list2 = sorted(number2, key=lambda a: a[0])

            above_k1 = [sorted_list1[i][1] for i in range(k)]
            above_k2 = [sorted_list2[i][1] for i in range(k)]
            forecast_label1 = max(set(above_k1), key=above_k1.count)
            forecast_label2 = max(set(above_k2), key=above_k2.count)

            true_label = v[1]
            if forecast_label1 == true_label:
                acc1 += 1
            if forecast_label2 == true_label:
                acc2 += 1



        accuracy1 = acc1 / len(validation) * 100
        accuracy2 = acc2 / len(validation) * 100
        print(f"L2 : {accuracy1:.2f}% , L1 : {accuracy2:.2f}%")
        sum_acc1.append(accuracy1)
        sum_acc2.append(accuracy2)

    mid2 = sum(sum_acc2)/5
    mid1 = sum(sum_acc1)/5
    center1.append(mid1)
    center2.append(mid2)

fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)

ax1.plot(K,center1,label='L1')
ax1.set_xlabel('K')
ax1.set_ylabel('midpoint Accuracy')
ax1.legend(loc='best')

ax2.plot(K,center2,label='L2')
ax2.set_xlabel('K')
ax2.set_ylabel('midpoint Accuracy')
ax2.legend(loc='best')


plt.tight_layout()
plt.show()
```
